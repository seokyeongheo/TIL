# _분류용 가상 데이터 생성 방법_

- make_classification
    - n_samples : 정수. 디폴트 100. 표본 데이터의 수.
    - n_features : 정수. 디폴트 20. 독립 변수의 수. 
    - n_informative : 정수. 디폴트 2. 독립 변수 중 종속 변수와 상관 관계가 있는 성분의 수.
    - n_redundant : 정수. 디폴트 2. 독립 변수 중 다른 독립 변수의 선형 조합으로 나타나는 성분의 수. 
    - n_repeated : 정수. 디폴트 0. 독립 변수 중 단순 중복된 성분의 수. 
    - n_classes : 정수. 디폴트 2. 종속 변수의 클래스 수. 
    - n_clusters per class : 정수. 디폴트 2. 클래스 당 클러스터의 수. 
    - weights : 리스트. 디폴트 NONE. 각 클래스에 할당된 표본 수. 
    - random_state : 정수. 디폴트 NONE. 난수 발생 시드. 
    

- make_blobs
    - n_samples : 정수. 디폴트 100. 표본 데이터의 수.
    - n_features : 정수. 디폴트 20. 독립 변수의 수.
    - n_cluters : 정수 또는 [n_center, n_feature] 크기의 배열. 디폴트 3. 생성할 클러스터의 수 혹은 중심.
    - cluster_std : 실수 혹은 실수열. 디폴트 1.0. 클러스터의 표준 편차. 
    - center_box : 실수쌍. 디폴트 (-10.0, 10.0) 생성할 클러스터의 바운딩 박스. 

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_features=1, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=4)
plt.title("One feature, One informative feature")
plt.scatter(X, y, marker='o', c=y, s=100, edgecolor="k", linewidth=1.5)
plt.show()
```

```python
X, y = make_classification(n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=4)
plt.title("One informative feature, One cluster per class")
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=100, edgecolor="k", linewidth=1.5)
plt.show()
```

```python
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=6)
plt.title("Two informative feature, One cluster per class")
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=60, edgecolor="k", linewidth=1.5)
plt.show()
```

```python
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, weights=[0.9, 0.1], random_state=6)
plt.title("Two informative feature, One cluster per class, different weight")
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=100, edgecolor='k', linewidth=1.5)
plt.show()
```

```python
X2, Y2 = make_classification(n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=2, random_state=6)
plt.title("Two informative feature, Two clusters per class")
plt.scatter(X2[: ,0], X2[:, 1], marker='o', c=Y2, s=100, edgecolor='k', linewidth=1.5)
plt.show()

# random_state=2, 
```

```python
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=3)
plt.title("Multi-class, Two informative features, One cluster")
plt.scatter(X[:, 0], X[:, 1], c=y, s=100, edgecolor='k', linewidth=1.5)
plt.show()
```

## 스케일링

```python
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
```

```python
x = (np.arange(9, dtype=np.float) - 3).reshape(-1, 1)
x = np.vstack([x, [20]])
df = pd.DataFrame(np.hstack([x, scale(x), robust_scale(x), minmax_scale(x), maxabs_scale(x)]), 
                  columns=["x", "scale(x)", "robust_scale(x)", 'minmax_scale(x)', "maxabs_scale(x)"])
df
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

data1 = iris.data
data2 = scale(iris.data)

print("old mean : ", np.mean(data1, axis=0))
print("old std : ", np.std(data1, axis=0))
print("new mean : ", np.mean(data2, axis=0))
print("new std : ", np.std(data2, axis=0))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(data1)
data2 = scaler.transform(data1)

data1.std(), data2.std()
```

## FunctionTransformer

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
def kernel(X):
    X0 = X[:, :1]
    X1 = X[:, 1:2]
    X2 = X[:, 2:3]
    X_new = np.hstack([X0, 2*X1, X2**2, np.log(X1)])
    return X_new
```

```python
X = np.arange(12).reshape(4, 3)
X
```

```python
kernel(X)
```

```python
FunctionTransformer(kernel).fit_transform(X)
```

## Label Encoder
- - -
- `classes` : 어떤 클래스가 있는지 확인
- `inverse_transform` : 다시 재변환

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le
```

```python
y = ['A', 'B', 'C', 'D', 'C', 'B', 'A', 'B', 'C']
y
```

```python
le.fit(y)
y2 = le.transform(y)
y2
```

```python
le.classes_
```

```python
le.inverse_transform(y2)
```
