```python
import numpy as np
import matplotlib.pyplot as plt
```

# _Gaussian Mixture_

```python
from numpy.random import randn

n_samples = 500
mu1 = np.array([0, 0])
mu2 = np.array([-6, 3])
sigma1 = np.array([[0., -0.1], [1.7, .4]])
sigma2 = np.eye(2)

np.random.seed(0)
X = np.r_[1.0 * np.dot(randn(n_samples, 2), sigma1) + mu1,
      0.7 * np.dot(randn(n_samples, 2), sigma2) + mu2,
     ]
plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], s=5)
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
```

```python
model = GaussianMixture(n_components=2, init_params='random', max_iter=1, random_state=0)
model.fit(X)
```

```python
gamma = model.predict_proba(X)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], s=50, linewidth=1, edgecolors='k', cmap=plt.cm.binary, c=gamma[:, 0])
plt.show()
```

```python
def plot_gaussianmixture(n):
    model = GaussianMixture(n_components=2, init_params='random', max_iter=n, random_state=0, tol=1e-9)
    model.fit(X)
    gamma = model.predict_proba(X)
    plt.figure(figsize=(10, 8))
    plt.title("Gaussian Mixture | Max iter : {}".format(n))
    plt.scatter(X[:, 0], X[:, 1], s=50, linewidth=1, edgecolors='k', cmap=plt.cm.binary, c=gamma[:, 0])
    plt.show()
```

```python
for i in range(5, 30, 5):
    plot_gaussianmixture(i)
```
