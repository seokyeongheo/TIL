```python
import numpy as np
```

```python
def make_regression(n_samples=10, n_features=2, n_target=1, bias=0.0, noise=0.0, random_state=0):
    
    np.random.seed(random_state)
    X = np.random.randn(n_samples, n_features)
    w = np.random.randn(n_features, 1)
    e = np.random.randn(n_target) * noise
    y = np.dot(X, w) + bias + e

    return X, y, w
```

```python
X, y, w = make_regression(n_samples=100, n_features=5, bias=3, noise=2, random_state=42)
print(X.shape, X, "\n")
print(w.shape, w, "\n")
print(y.shape, y)
```
