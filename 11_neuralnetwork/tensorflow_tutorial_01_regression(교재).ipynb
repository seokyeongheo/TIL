```python
import tensorflow as tf
import numpy as np
```

### multi-variable linear regression

```python
# placeholder, variable

x_data = np.array([[73., 80., .75], [93., 88., 93.], [89., 91., 90.], 
                   [96., 98., 100.], [73., 66., 70.]])
y_data = np.array([[152.], [185.], [180.], [196.], [142.]])
x_data.shape, y_data.shape, type(x_data), type(y_data)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

Y_ = tf.matmul(X, W) + b
```

```python
# cost function, optimizer, modeling

cost = tf.reduce_mean(tf.square(Y_ - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
# launch

with tf.Session() as sess:
    
    init = tf.global_variables_initializer()
    sess.run(init)
    
    for step in range(2001):
        cost_val, hy_val, _ = sess.run([cost, Y_, train], feed_dict={X: x_data, Y: y_data})
        
        if step % 10 == 0:
            print(step, "Cost: ", cost_val, "Prediction: ", hy_val)
```
