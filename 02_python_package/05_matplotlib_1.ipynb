```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### basic line plot

```python
y = np.random.randint(1, 10, size=10)
plt.plot(y)
plt.show()
```

```python
# x  

x = range(0, 100, 10)
plt.plot(x, y)
plt.show()
```

### 

- option
    - color
    - marker
    - line
- custom style

```python
# color maker

plt.plot(y, 'o-')
plt.show()
```

```python
# custom style

# ms : marker size
# mec : marker color
# mew : marker width 
# c : line color / lw : line  / lw :  

plt.plot(y, lw = 3, ls = "-", marker = "o", ms = 5, mew = 1, mfc = "lightpink") 
plt.show()
```

```python
#### limit  

plt.plot(y)
plt.xlim(-1, 10) # start ~ end.   x 
plt.ylim(-5, 10)
plt.show()
```

```python
# tic 

plt.plot(y)
plt.xticks(range(10))
plt.yticks([1,2,3,4,5])

plt.show()
```

```python
# grid 

plt.plot(y)
plt.grid(True)
plt.show()
```

```python
#  (multi line) 

y1 = np.random.randint(1, 10, size = 5)
y2 = np.random.randint(1, 10, size = 5)
y3 = np.random.randint(1, 10, size = 5)
x = range(5)

print(list(x), y1, y2, y3)
plt.plot(x, y1, 'o-', x, y2, '*-.', x, y3, '^:')
plt.show()
```

```python
#    

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

### 

- label, location

```python
y1 = np.random.randint(1, 10, size = 5)
y2 = np.random.randint(1, 10, size = 5)

plt.plot(y1, label = "y1")
plt.plot(y2, label = "y2")
plt.legend(loc = 0)
plt.show()
```

### title 

```python
plt.plot(y1)
plt.title("Title")
plt.xlabel("x ")
plt.ylabel("ylabel")
plt.show()
```
