```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### basic line plot

```python
y = np.random.randint(1, 10, size=10)
plt.plot(y)
plt.show()
```

```python
# x 데이터 추가

x = range(0, 100, 10)
plt.plot(x, y)
plt.show()
```

### 스타일

- option
    - color
    - marker
    - line
- custom style

```python
# color maker

plt.plot(y, 'o-')
plt.show()
```

```python
# custom style

# ms : marker size
# mec : marker color
# mew : marker width 
# c : line color / lw : line 두께 / lw : 선 모양

plt.plot(y, lw = 3, ls = "-", marker = "o", ms = 5, mew = 1, mfc = "lightpink") 
plt.show()
```

```python
#### limit 설정 

plt.plot(y)
plt.xlim(-1, 10) # start ~ end. 이 안에서 x가 움직이도록
plt.ylim(-5, 10)
plt.show()
```

```python
# tic 설정

plt.plot(y)
plt.xticks(range(10))
plt.yticks([1,2,3,4,5])

plt.show()
```

```python
# grid 추가

plt.plot(y)
plt.grid(True)
plt.show()
```

```python
# 다중 데이터(multi line) 표현

y1 = np.random.randint(1, 10, size = 5)
y2 = np.random.randint(1, 10, size = 5)
y3 = np.random.randint(1, 10, size = 5)
x = range(5)

print(list(x), y1, y2, y3)
plt.plot(x, y1, 'o-', x, y2, '*-.', x, y3, '^:')
plt.show()
```

```python
# 다중 데이터 구혀 

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

### 범례

- label, location

```python
y1 = np.random.randint(1, 10, size = 5)
y2 = np.random.randint(1, 10, size = 5)

plt.plot(y1, label = "y1")
plt.plot(y2, label = "y2")
plt.legend(loc = 0)
plt.show()
```

### title 추가

```python
plt.plot(y1)
plt.title("Title")
plt.xlabel("x 축")
plt.ylabel("ylabel")
plt.show()
```

### Figure

- 그래프가 그려지는 영역 
- 하나의 figure에 여러 개의 그래프를 그릴 수 있다
    - 그래프의 크기를 조절
    - 한꺼번에 여러 개의 그래프 그리기

```python
# 그래프의 크기 조절

plt.figure(figsize = (12, 6)) # (가로, 세로)
plt.plot(y)
plt.plot(y1)
plt.show()
```

```python
# 여러 개의 그래프 그리기

print(y, y1, y2, y3)

plt.figure(figsize = (20, 10)) # (가로, 세로)

y4 = y

plt.subplot(2, 2, 1) # two by one matrix의 1의 자리에 그래프 그리기
plt.grid()
plt.plot(y)

plt.subplot(2, 2, 2) # two by one matrix의 1의 자리에 그래프 그리기
plt.grid()
plt.plot(y1)

plt.subplot(2, 2, 3) # two by one matrix의 2의 자리에 그래프 그리기
plt.grid()
plt.plot(y2)

plt.subplot(2, 2, 4) # two by one matrix의 1의 자리에 그래프 그리기
plt.grid()
plt.plot(y4)

plt.show()
```
