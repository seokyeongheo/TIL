## *matplotlib2*

- bar chart
- pie chart
- scatter plot
- histogram

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### bar chart

```python
y = np.random.randint(1, 10, size=10)
x = range(10)
plt.bar(x, y)
plt.show()
```

```python
# bar plot에 error 표시하기
# 투명도 적용하기 alpha = 
# bar graph 수평으로 그리기

y = np.random.randint(1, 10, size=10)
x = range(10)
error = np.random.randint(1, 3, size=10)
plt.barh(x, y, xerr = error, alpha = 0.5)
plt.show()
```

```python
y1 = np.random.randint(6, 10, size=10)
y2 = np.random.randint(1, 5, size=10)
x = range(10)

plt.barh(x, y1, alpha = 0.7)
plt.barh(x, y2, alpha = 0.7)
plt.show()
```

### pie chart

```python
# label 적용하기 {labels = }
# % 적용하기 autopct = "%1.2f%" %1 >> 자연수를 넣는다 // .2f% >> 소수 둘째짜리 까지 남긴다 // %d%% 10진수로 표현한다.
# 색상 변경하기

y = np.random.randint(6, 10, size=5)
labels = ["Canada", "America", "Korea", "Japan", "Germany"]
colors = ["yellow", "gold", "lightskyblue", 'lightcoral', "r"]

plt.pie(y, labels=labels, colors=colors, autopct='%df%%')
plt.show()
```

### scatter chart

- numpy random rand(uniform), randn(정규분포), randint(균등분포). scatter plot으로 확인 가능하다.

```python
# random.rand 확인

y = np.random.rand(500)
x = np.random.rand(500)
plt.scatter(x, y)
plt.show()
```

```python
# random.randn 확인

y = np.random.randn(500)
x = np.random.randn(500)
plt.scatter(x, y)
plt.show()
```

```python
# random.randint 확인

y = np.random.randint(1, 100, 500)
x = np.random.randint(1, 100, 500)
plt.scatter(x, y)
plt.show()
```

```python
# 4차원 데이터(x, y, 색, 크기)를 scatter로 나타내기

y = np.random.randint(1, 10, size=20)
x = np.random.randint(1, 10, size=20)
colors = np.random.randint(1, 10, size=20)
area = np.pi * np.random.randint(10, 30, size=20)**2

plt.scatter(y, x, c=colors, s=area, alpha=0.3)
plt.show()
```

### histogram

```python
# 특정 x범위에서 random하게 나온 개체 수를 y값에 출력해준 것.

y = np.random.randint(1, 100, size=100)

plt.hist(y, bins=100)
plt.show()
```
