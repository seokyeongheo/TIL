## _Seaborn_

- matplotlib를 기반으로 더욱 고급스러운 인터페이스를 제공하기 위해 개발된 python 시각화 라이브러리

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def sinplot():
    x = np.linspace(0, 10, 101)
    for i in range(6):
        plt.plot(x, np.sin(i + x))
sinplot()
```

```python
# import된 seaborn을 사용하도록. 

sns.set()
```

```python
sinplot()
```

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("darkgrid")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
# 테두리 선 조정

sinplot()
sns.despine()
```

```python
# context 설정 - 그래프의 굵기
# paper, notebook, talk, poster
```

```python
sns.set_context("paper")
sinplot()
```

```python
sns.set_context("notebook")
sinplot()
```

```python
sns.set_context("talk")
sinplot()
```

```python
sns.set_context("poster")
sinplot()
```

### color palettes

- 색상 테마 설정이 가능

```python
# 현재 사용 중인 색상 테마

cp = sns.color_palette()
sns.palplot(cp)
```

```python
p = sns.color_palette("hls", 8)
sns.palplot(p)
```

```python
# palette 적용 

sns.set_palette("hls", 8)
sinplot()
```

```python
# 테마 확인

colors = ["muted", "hls", "RdBu", "CMRmap", "rainbow"]
for color in colors:
    p = sns.color_palette(color, 8)
    sns.palplot(p)
```

### sample data

```python
iris = sns.load_dataset("iris")
iris.tail()
```

```python
# histogram

sns.rugplot(iris.petal_length.values)
```

```python
# kde plot

sns.kdeplot(iris.petal_length.values)
```

```python
#distplot

sns.distplot(iris.petal_length.values, rug = True)
```

### joint plot

- 2개의 column data를 나타내주는 plot. 

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris)
```

### pair plot

```python
# 모든 columns 쌍을 scatter plot으로 보여준다
# 전체 흐름을 한 눈에 파악할 수 있는 함수(?)

sns.pairplot(iris, hue="species")
plt.show()
```

### heatmap

```python
flight = sns.load_dataset("flights")
flight_data = flight.pivot("month", "year", "passengers")
flight_data
```

```python
sns.set_context("poster")
```

```python
# cmap 등등 : 테마 설정
# annot = annotation? 글자 추가. 
# fmt = "d"(자연수만 넣을 수 있도록 하기)

plt.figure(figsize = (20, 15)) # (가로, 세로)
sns.heatmap(flight_data, cmap = "YlGnBu", annot = True, fmt = "d")
```
