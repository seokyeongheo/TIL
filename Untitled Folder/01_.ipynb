```python
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.set()
```

```python
# 연습문제 1 - 1
# 0, 1, 9, 10번 나오는 경우에 동전이 fair하지 않다. 
# ( 4 / 11 = 0.363636...)

for i in range(0, 11):
    print(i, "번 나오는 유의 확률은", round(stats.binom_test(i, 10), 4))
```

```python
# 연습문제 1 - 2
# 473번 이하, 527번 이상 나오는 경우에 동전이 fair하지 않다. 
# ( 948 / 1001 = 0.94705...)

for i in range(0, 1001):
    print(i, "번 나오는 유의 확률은", round(stats.binom_test(i, 1000), 4))
```

```python
# 연습문제 3

tips = sns.load_dataset("tips")
tips.tail()
```

```python
tips_df = tips.groupby(["sex", "smoker"]).size().reset_index(name='counts')
tips_df
```

```python
# "여자 손님 중 비흡연자가 흡연자보다 많다"라는 명제를 맞다고 할 수 있다.

stats.binom_test(54, 87)
```

```python
x = [0, 1]
y = tips_df['counts'].values

plt.bar(x, y[2:])
plt.show()
```

```python
tips_df2 = tips.groupby(["sex", "smoker", "time"]).size().reset_index(name='counts')
tips_df2
```

```python
# "저녁에 오는 여자 손님 중 비흡연자가 흡연자보다 많다"라는 명제가 아니라고 할 수 없다.

stats.binom_test(29, 52)
```
